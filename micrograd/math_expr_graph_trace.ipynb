```python
from micrograd.viz import draw
from micrograd.value import Value
```

```python
# simple example

x = Value(1.0)
y = (x * 2 + 1).relu()
y.backward()
draw(y)
```

```python
# simple 2D neuron
import random
from micrograd.nn import Neuron

random.seed(1337)
n = Neuron(2)
x = [Value(1.0), Value(-2.0)]
y = n(x)
y.backward()

dot = draw(y)
dot
```
